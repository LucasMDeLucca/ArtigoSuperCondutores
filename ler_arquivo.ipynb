```python
"""
Script para ler arquivo HDF5 completo e converter para DataFrame do pandas
Baseado no arquivo 0724.hdf5 de 12.38 GB com dados de supercondutores

Autor: Assistente IA
Data: 2025-08-30
"""

import h5py
import pandas as pd
import numpy as np
import pickle
import warnings
from typing import Dict, List, Any, Union
import sys
from tqdm import tqdm
import gc

# Configurações
warnings.filterwarnings('ignore')

def decode_object_data(data: Any) -> Any:
    """
    Decodifica dados do tipo object que podem ser strings em bytes ou objetos pickle
    
    Args:
        data: Dados a serem decodificados
        
    Returns:
        Dados decodificados
    """
    if isinstance(data, np.ndarray):
        if data.dtype.kind == 'O':  # object dtype
            decoded_list = []
            for item in data:
                if isinstance(item, (bytes, bytearray)):
                    try:
                        # Tenta decodificar como string UTF-8
                        decoded_list.append(item.decode('utf-8'))
                    except UnicodeDecodeError:
                        try:
                            # Tenta decodificar como pickle
                            decoded_list.append(pickle.loads(item))
                        except:
                            # Se falhar, mantém como bytes
                            decoded_list.append(item)
                else:
                    decoded_list.append(item)
            return decoded_list
        else:
            return data.tolist() if hasattr(data, 'tolist') else data
    else:
        return data

def flatten_array(arr: np.ndarray, prefix: str = '') -> Dict[str, Any]:
    """
    Achata arrays multidimensionais em colunas separadas
    
    Args:
        arr: Array numpy a ser achatado
        prefix: Prefixo para os nomes das colunas
        
    Returns:
        Dicionário com dados achatados
    """
    if arr.ndim == 1:
        if len(arr) == 1:
            return {prefix: arr[0]}
        else:
            return {f"{prefix}_{i}": arr[i] for i in range(len(arr))}
    elif arr.ndim == 2:
        result = {}
        for i in range(arr.shape[0]):
            for j in range(arr.shape[1]):
                result[f"{prefix}_{i}_{j}"] = arr[i, j]
        return result
    elif arr.ndim == 3:
        # Para arrays 3D, pegamos apenas algumas estatísticas para evitar explosão de colunas
        result = {
            f"{prefix}_shape_0": arr.shape[0],
            f"{prefix}_shape_1": arr.shape[1],
            f"{prefix}_shape_2": arr.shape[2],
            f"{prefix}_mean": np.mean(arr),
            f"{prefix}_std": np.std(arr),
            f"{prefix}_min": np.min(arr),
            f"{prefix}_max": np.max(arr)
        }
        return result
    elif arr.ndim >= 4:
        # Para arrays 4D+, apenas informações básicas
        result = {
            f"{prefix}_ndim": arr.ndim,
            f"{prefix}_size": arr.size,
            f"{prefix}_mean": np.mean(arr),
            f"{prefix}_std": np.std(arr),
            f"{prefix}_min": np.min(arr),
            f"{prefix}_max": np.max(arr)
        }
        for i, dim_size in enumerate(arr.shape):
            result[f"{prefix}_shape_{i}"] = dim_size
        return result
    else:
        return {prefix: arr}

def process_dataset(dataset: h5py.Dataset, dataset_name: str) -> Dict[str, Any]:
    """
    Processa um dataset individual do HDF5
    
    Args:
        dataset: Dataset do HDF5
        dataset_name: Nome do dataset
        
    Returns:
        Dicionário com dados processados
    """
    try:
        # Carrega os dados
        data = dataset[()]
        
        # Processa baseado no tipo e forma dos dados
        if dataset.dtype.kind == 'O':  # object dtype
            decoded_data = decode_object_data(data)
            if isinstance(decoded_data, list):
                # Se for uma lista de strings/objetos, cria colunas separadas
                result = {}
                for i, item in enumerate(decoded_data):
                    result[f"{dataset_name}_{i}"] = str(item)
                return result
            else:
                return {dataset_name: str(decoded_data)}
        
        elif isinstance(data, np.ndarray):
            return flatten_array(data, dataset_name)
        
        else:
            return {dataset_name: data}
            
    except Exception as e:
        print(f"Erro ao processar dataset {dataset_name}: {e}")
        return {f"{dataset_name}_error": str(e)}

def read_hdf5_to_dataframe(file_path: str, max_groups: int = None) -> pd.DataFrame:
    """
    Lê arquivo HDF5 completo e converte para DataFrame do pandas
    
    Args:
        file_path: Caminho para o arquivo HDF5
        max_groups: Número máximo de grupos a processar (None para todos)
        
    Returns:
        DataFrame do pandas com todos os dados
    """
    print(f"Abrindo arquivo HDF5: {file_path}")
    
    all_data = []
    
    with h5py.File(file_path, 'r') as f:
        # Lista todos os grupos top-level
        groups = list(f.keys())
        print(f"Encontrados {len(groups)} grupos no arquivo")
        
        if max_groups:
            groups = groups[:max_groups]
            print(f"Processando apenas os primeiros {max_groups} grupos")
        
        # Processa cada grupo
        for group_idx, group_name in enumerate(tqdm(groups, desc="Processando grupos")):
            try:
                group = f[group_name]
                row_data = {'group_id': group_name}
                
                # Processa todos os datasets dentro do grupo
                for dataset_name in group.keys():
                    if isinstance(group[dataset_name], h5py.Dataset):
                        dataset_data = process_dataset(group[dataset_name], dataset_name)
                        row_data.update(dataset_data)
                
                all_data.append(row_data)
                
                # Libera memória periodicamente
                if (group_idx + 1) % 100 == 0:
                    gc.collect()
                    
            except Exception as e:
                print(f"Erro ao processar grupo {group_name}: {e}")
                continue
    
    print("Criando DataFrame...")
    df = pd.DataFrame(all_data)
    
    print(f"DataFrame criado com {len(df)} linhas e {len(df.columns)} colunas")
    print(f"Uso de memória: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    
    return df

def save_dataframe(df: pd.DataFrame, output_path: str, format: str = 'csv'):
    """
    Salva o DataFrame em diferentes formatos
    
    Args:
        df: DataFrame a ser salvo
        output_path: Caminho de saída (sem extensão)
        format: Formato de saída ('csv', 'parquet', 'pickle', 'excel')
    """
    if format == 'csv':
        df.to_csv(f"{output_path}.csv", index=False)
        print(f"DataFrame salvo como CSV: {output_path}.csv")
    
    elif format == 'parquet':
        df.to_parquet(f"{output_path}.parquet", index=False)
        print(f"DataFrame salvo como Parquet: {output_path}.parquet")
    
    elif format == 'pickle':
        df.to_pickle(f"{output_path}.pkl")
        print(f"DataFrame salvo como Pickle: {output_path}.pkl")
    
    elif format == 'excel':
        # Excel tem limite de linhas, então vamos dividir se necessário
        max_rows = 1048576  # Limite do Excel
        if len(df) > max_rows:
            print(f"DataFrame tem {len(df)} linhas, dividindo em múltiplos arquivos Excel...")
            for i in range(0, len(df), max_rows):
                chunk = df.iloc[i:i+max_rows]
                chunk.to_excel(f"{output_path}_part_{i//max_rows + 1}.xlsx", index=False)
                print(f"Parte {i//max_rows + 1} salva: {output_path}_part_{i//max_rows + 1}.xlsx")
        else:
            df.to_excel(f"{output_path}.xlsx", index=False)
            print(f"DataFrame salvo como Excel: {output_path}.xlsx")

def main():
    """
    Função principal
    """
    # Configurações
    file_path = "0724.hdf5"  # Altere para o caminho correto do seu arquivo
    output_path = "dados_hdf5_completo"
    
    # Opções de processamento
    max_groups = None  # None para processar todos os grupos, ou um número para limitar
    
    print("=== LEITOR DE ARQUIVO HDF5 COMPLETO ===")
    print(f"Arquivo: {file_path}")
    
    try:
        # Lê o arquivo HDF5 completo
        df = read_hdf5_to_dataframe(file_path, max_groups)
        
        # Mostra informações básicas
        print("\n=== INFORMAÇÕES DO DATAFRAME ===")
        print(f"Dimensões: {df.shape}")
        print(f"Colunas: {list(df.columns)[:10]}...")  # Mostra apenas as primeiras 10 colunas
        print(f"Tipos de dados únicos: {df.dtypes.value_counts().to_dict()}")
        
        # Mostra amostra dos dados
        print("\n=== AMOSTRA DOS DADOS ===")
        print(df.head())
        
        # Salva em diferentes formatos
        print("\n=== SALVANDO DATAFRAME ===")
        save_dataframe(df, output_path, 'csv')
        save_dataframe(df, output_path, 'parquet')
        save_dataframe(df, output_path, 'pickle')
        
        print("\n=== PROCESSAMENTO CONCLUÍDO ===")
        print(f"DataFrame final: {df.shape[0]} linhas x {df.shape[1]} colunas")
        
        return df
        
    except Exception as e:
        print(f"Erro durante o processamento: {e}")
        import traceback
        traceback.print_exc()
        return None

if __name__ == "__main__":
    df = main()
```

```python
"""
Script otimizado para ler arquivos HDF5 muito grandes (12GB+) e converter para DataFrame
Versão com processamento em chunks e otimizações de memória

Autor: Assistente IA
Data: 2025-08-30
"""

import h5py
import pandas as pd
import numpy as np
import pickle
import warnings
from typing import Dict, List, Any, Union, Optional
import sys
import os
from tqdm import tqdm
import gc
import psutil
from concurrent.futures import ThreadPoolExecutor, as_completed
import time

# Configurações
warnings.filterwarnings('ignore')

class HDF5Reader:
    """
    Classe para leitura otimizada de arquivos HDF5 grandes
    """
    
    def __init__(self, file_path: str, chunk_size: int = 1000, max_memory_gb: float = 4.0):
        """
        Inicializa o leitor HDF5
        
        Args:
            file_path: Caminho para o arquivo HDF5
            chunk_size: Número de grupos a processar por vez
            max_memory_gb: Máximo de memória a usar em GB
        """
        self.file_path = file_path
        self.chunk_size = chunk_size
        self.max_memory_gb = max_memory_gb
        self.processed_groups = 0
        self.total_groups = 0
        
    def get_memory_usage(self) -> float:
        """Retorna o uso atual de memória em GB"""
        process = psutil.Process(os.getpid())
        return process.memory_info().rss / 1024**3
    
    def decode_object_data(self, data: Any) -> Any:
        """Decodifica dados do tipo object"""
        if isinstance(data, np.ndarray):
            if data.dtype.kind == 'O':
                decoded_list = []
                for item in data:
                    if isinstance(item, (bytes, bytearray)):
                        try:
                            decoded_list.append(item.decode('utf-8'))
                        except UnicodeDecodeError:
                            try:
                                decoded_list.append(str(pickle.loads(item)))
                            except:
                                decoded_list.append(str(item))
                    else:
                        decoded_list.append(str(item))
                return decoded_list
            else:
                return data.tolist() if hasattr(data, 'tolist') else data
        else:
            return data
    
    def process_large_array(self, arr: np.ndarray, name: str) -> Dict[str, Any]:
        """
        Processa arrays grandes extraindo apenas estatísticas essenciais
        """
        result = {
            f"{name}_shape": str(arr.shape),
            f"{name}_dtype": str(arr.dtype),
            f"{name}_size": arr.size
        }
        
        # Para arrays muito grandes, apenas estatísticas básicas
        if arr.size > 100000:  # Mais de 100k elementos
            try:
                # Amostra aleatória para estatísticas
                if arr.size > 1000000:
                    sample_size = min(10000, arr.size // 100)
                    flat_arr = arr.flatten()
                    indices = np.random.choice(len(flat_arr), sample_size, replace=False)
                    sample = flat_arr[indices]
                else:
                    sample = arr.flatten()
                
                if np.issubdtype(arr.dtype, np.number):
                    result.update({
                        f"{name}_mean": float(np.mean(sample)),
                        f"{name}_std": float(np.std(sample)),
                        f"{name}_min": float(np.min(sample)),
                        f"{name}_max": float(np.max(sample))
                    })
            except:
                result[f"{name}_stats_error"] = "Erro ao calcular estatísticas"
        else:
            # Para arrays menores, pode incluir mais detalhes
            if arr.ndim <= 2 and arr.size <= 1000:
                return self.flatten_small_array(arr, name)
            else:
                try:
                    if np.issubdtype(arr.dtype, np.number):
                        result.update({
                            f"{name}_mean": float(np.mean(arr)),
                            f"{name}_std": float(np.std(arr)),
                            f"{name}_min": float(np.min(arr)),
                            f"{name}_max": float(np.max(arr))
                        })
                except:
                    pass
        
        return result
    
    def flatten_small_array(self, arr: np.ndarray, prefix: str) -> Dict[str, Any]:
        """Achata arrays pequenos em colunas separadas"""
        if arr.ndim == 1:
            if len(arr) <= 10:  # Limite para evitar muitas colunas
                return {f"{prefix}_{i}": arr[i] for i in range(len(arr))}
            else:
                return {f"{prefix}_array": str(arr.tolist())}
        
        elif arr.ndim == 2:
            if arr.size <= 50:  # Limite para matrizes pequenas
                result = {}
                for i in range(arr.shape[0]):
                    for j in range(arr.shape[1]):
                        result[f"{prefix}_{i}_{j}"] = arr[i, j]
                return result
            else:
                return {f"{prefix}_matrix": str(arr.tolist())}
        
        else:
            return {f"{prefix}_array": str(arr.shape)}
    
    def process_dataset_optimized(self, dataset: h5py.Dataset, dataset_name: str) -> Dict[str, Any]:
        """
        Processa um dataset de forma otimizada
        """
        try:
            # Informações básicas do dataset
            result = {
                f"{dataset_name}_shape": str(dataset.shape),
                f"{dataset_name}_dtype": str(dataset.dtype),
                f"{dataset_name}_size": dataset.size
            }
            
            # Decide como processar baseado no tamanho
            if dataset.size == 0:
                result[f"{dataset_name}_empty"] = True
                return result
            
            # Para datasets muito grandes, não carrega tudo na memória
            if dataset.size > 1000000:  # Mais de 1M elementos
                try:
                    # Amostra apenas uma pequena parte
                    if dataset.ndim == 1:
                        sample = dataset[:min(1000, len(dataset))]
                    elif dataset.ndim == 2:
                        sample = dataset[:min(100, dataset.shape[0]), :min(100, dataset.shape[1])]
                    else:
                        # Para arrays multidimensionais, pega apenas uma fatia
                        slices = tuple(slice(0, min(10, dim)) for dim in dataset.shape)
                        sample = dataset[slices]
                    
                    if dataset.dtype.kind == 'O':
                        decoded = self.decode_object_data(sample)
                        result[f"{dataset_name}_sample"] = str(decoded)[:200]  # Limita o tamanho
                    else:
                        result.update(self.process_large_array(sample, f"{dataset_name}_sample"))
                        
                except Exception as e:
                    result[f"{dataset_name}_sample_error"] = str(e)
            
            else:
                # Para datasets menores, carrega completamente
                data = dataset[()]
                
                if dataset.dtype.kind == 'O':
                    decoded_data = self.decode_object_data(data)
                    if isinstance(decoded_data, list) and len(decoded_data) <= 20:
                        for i, item in enumerate(decoded_data):
                            result[f"{dataset_name}_{i}"] = str(item)
                    else:
                        result[f"{dataset_name}_data"] = str(decoded_data)[:500]
                
                elif isinstance(data, np.ndarray):
                    result.update(self.process_large_array(data, dataset_name))
                
                else:
                    result[dataset_name] = data
            
            return result
            
        except Exception as e:
            return {f"{dataset_name}_error": str(e)}
    
    def process_group(self, group_name: str, file_handle: h5py.File) -> Dict[str, Any]:
        """
        Processa um grupo individual
        """
        try:
            group = file_handle[group_name]
            row_data = {'group_id': group_name}
            
            # Processa todos os datasets dentro do grupo
            for dataset_name in group.keys():
                if isinstance(group[dataset_name], h5py.Dataset):
                    dataset_data = self.process_dataset_optimized(group[dataset_name], dataset_name)
                    row_data.update(dataset_data)
            
            return row_data
            
        except Exception as e:
            return {'group_id': group_name, 'error': str(e)}
    
    def read_in_chunks(self, max_groups: Optional[int] = None, 
                      output_prefix: str = "chunk", 
                      save_chunks: bool = True) -> pd.DataFrame:
        """
        Lê o arquivo HDF5 em chunks para otimizar o uso de memória
        """
        print(f"Iniciando leitura em chunks do arquivo: {self.file_path}")
        
        all_dataframes = []
        chunk_files = []
        
        with h5py.File(self.file_path, 'r') as f:
            groups = list(f.keys())
            self.total_groups = len(groups)
            
            if max_groups:
                groups = groups[:max_groups]
                print(f"Limitando a {max_groups} grupos")
            
            print(f"Total de grupos a processar: {len(groups)}")
            
            # Processa em chunks
            for chunk_idx in range(0, len(groups), self.chunk_size):
                chunk_groups = groups[chunk_idx:chunk_idx + self.chunk_size]
                chunk_data = []
                
                print(f"\nProcessando chunk {chunk_idx//self.chunk_size + 1} "
                      f"({len(chunk_groups)} grupos)...")
                
                # Processa grupos do chunk atual
                for group_name in tqdm(chunk_groups, desc=f"Chunk {chunk_idx//self.chunk_size + 1}"):
                    row_data = self.process_group(group_name, f)
                    chunk_data.append(row_data)
                    self.processed_groups += 1
                    
                    # Verifica uso de memória
                    if self.get_memory_usage() > self.max_memory_gb:
                        print(f"Limite de memória atingido: {self.get_memory_usage():.2f}GB")
                        gc.collect()
                
                # Cria DataFrame do chunk
                chunk_df = pd.DataFrame(chunk_data)
                
                if save_chunks:
                    # Salva chunk individual
                    chunk_file = f"{output_prefix}_chunk_{chunk_idx//self.chunk_size + 1}.parquet"
                    chunk_df.to_parquet(chunk_file, index=False)
                    chunk_files.append(chunk_file)
                    print(f"Chunk salvo: {chunk_file}")
                    
                    # Libera memória
                    del chunk_data
                    gc.collect()
                else:
                    all_dataframes.append(chunk_df)
                
                print(f"Memória atual: {self.get_memory_usage():.2f}GB")
        
        # Combina todos os chunks
        if save_chunks:
            print("\nCombinando chunks...")
            combined_df = pd.concat([pd.read_parquet(f) for f in chunk_files], 
                                  ignore_index=True)
            
            # Remove arquivos temporários se desejado
            # for f in chunk_files:
            #     os.remove(f)
            
        else:
            combined_df = pd.concat(all_dataframes, ignore_index=True)
        
        print(f"\nDataFrame final: {combined_df.shape}")
        return combined_df
    
    def get_file_info(self) -> Dict[str, Any]:
        """
        Obtém informações básicas sobre o arquivo HDF5
        """
        info = {}
        
        with h5py.File(self.file_path, 'r') as f:
            info['total_groups'] = len(f.keys())
            info['group_names'] = list(f.keys())[:10]  # Primeiros 10
            
            # Analisa estrutura de alguns grupos
            sample_groups = list(f.keys())[:5]
            datasets_info = {}
            
            for group_name in sample_groups:
                group = f[group_name]
                datasets_info[group_name] = {}
                
                for dataset_name in group.keys():
                    if isinstance(group[dataset_name], h5py.Dataset):
                        ds = group[dataset_name]
                        datasets_info[group_name][dataset_name] = {
                            'shape': ds.shape,
                            'dtype': str(ds.dtype),
                            'size': ds.size
                        }
            
            info['sample_datasets'] = datasets_info
        
        return info

def main():
    """
    Função principal otimizada
    """
    # Configurações
    file_path = "0724.hdf5"  # Altere para o caminho correto
    output_prefix = "dados_hdf5_otimizado"
    
    # Parâmetros de otimização
    chunk_size = 500  # Grupos por chunk
    max_memory_gb = 4.0  # Máximo de memória a usar
    max_groups = None  # None para todos, ou número para limitar
    
    print("=== LEITOR HDF5 OTIMIZADO PARA ARQUIVOS GRANDES ===")
    
    # Inicializa o leitor
    reader = HDF5Reader(file_path, chunk_size, max_memory_gb)
    
    # Mostra informações do arquivo
    print("\n=== INFORMAÇÕES DO ARQUIVO ===")
    try:
        file_info = reader.get_file_info()
        print(f"Total de grupos: {file_info['total_groups']}")
        print(f"Primeiros grupos: {file_info['group_names']}")
        print("Estrutura de datasets (amostra):")
        for group, datasets in list(file_info['sample_datasets'].items())[:2]:
            print(f"  {group}:")
            for ds_name, ds_info in datasets.items():
                print(f"    {ds_name}: {ds_info['shape']} ({ds_info['dtype']})")
    except Exception as e:
        print(f"Erro ao obter informações: {e}")
    
    # Processa o arquivo
    print(f"\n=== INICIANDO PROCESSAMENTO ===")
    start_time = time.time()
    
    try:
        df = reader.read_in_chunks(max_groups=max_groups, 
                                 output_prefix=output_prefix,
                                 save_chunks=True)
        
        end_time = time.time()
        processing_time = end_time - start_time
        
        print(f"\n=== RESULTADO FINAL ===")
        print(f"Tempo de processamento: {processing_time:.2f} segundos")
        print(f"DataFrame final: {df.shape[0]} linhas x {df.shape[1]} colunas")
        print(f"Uso de memória: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
        
        # Salva resultado final
        final_output = f"{output_prefix}_final.parquet"
        df.to_parquet(final_output, index=False)
        print(f"Resultado salvo: {final_output}")
        
        # Também salva em CSV se não for muito grande
        if len(df) < 500000:  # Menos de 500k linhas
            df.to_csv(f"{output_prefix}_final.csv", index=False)
            print(f"CSV salvo: {output_prefix}_final.csv")
        
        # Mostra amostra
        print("\n=== AMOSTRA DOS DADOS ===")
        print(df.head())
        print(f"\nColunas ({len(df.columns)}): {list(df.columns)[:10]}...")
        
        return df
        
    except Exception as e:
        print(f"Erro durante o processamento: {e}")
        import traceback
        traceback.print_exc()
        return None

if __name__ == "__main__":
    df = main()
```
