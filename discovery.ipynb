## Ajustando as colunas com o do tipo shape

```python
df.filter(like = 'shape').head()
```

```python
colunas_criadas = []
for i in df.filter(like = 'shape').columns:
    colunas = []
    qtd_termos = df[i].str.count(',').values[0]+1

    print(f'Usando a coluna {i}, criamos as colunas: ')

    for j in range(1, int(qtd_termos+1)):
         print(i+'_'+str(j))
         colunas.append(i+'_'+str(j))

    colunas_criadas += colunas

    df[colunas] = (
    df[i]
    .str.strip("()")      # remove os parênteses
    .str.split(",", expand=True)  # separa pelo separador vírgula
    )

    features_retiradas_individuais.append(i)

    print("------------------//---------------\n")
    
    
```

```python
colunas_extraidas = features_num_retiradas+features_cat_retiradas+features_retiradas_individuais
```

# Verificacao de novas colunas criadas

```python
def check_distribuition(df, col):
    qtd_infos = len(df[col].value_counts(dropna = False).index.values)

    qtd_nulos = (df.shape[0] - qtd_infos)/df.shape[0]

    if  qtd_nulos >= 0.95:
        df.drop(columns = [col], inplace=True)
        print(f'Coluna {col} retirada por conter {int(100*qtd_nulos)}% de nulos')

    return df
```

```python
for i in colunas_criadas:
    df_cleaned_1 = check_distribuition(df_cleaned_1, i)
```
